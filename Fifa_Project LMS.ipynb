```python
# creating new features by aggregating the features

def defend(fifa_data):
    return int(round((fifa_data[['Marking', 'StandingTackle',
                                 'SlidingTackle']].mean()).mean()))

def general(fifa_data):
    return int(round((fifa_data[['HeadingAccuracy', 'Dribbling', 'Curve',
                                'BallControl']].mean()).mean()))

def mental(fifa_data):
    return int(round((fifa_data[['Aggression', 'Interceptions', 'Positioning', 'Vision',
                                'Composure']].mean()).mean()))

def passing(fifa_data):
    return int(round((fifa_data[['Crossing','ShortPassing', 'LongPassing']].mean()).mean()))


def mobility(fifa_data):
    return int(round((fifa_data[['Acceleration','SprintSpeed', 'Agility',
                                 'Reactions']].mean()).mean()))

def power(fifa_data):
    return int(round((fifa_data[['Balance', 'Jumping','Stamina','Strength']].mean()).mean()))


def rating(fifa_data):
    return int(round((fifa_data[['Potential', 'Overall']].mean()).mean()))


def shooting(fifa_data):
    return int(round((fifa_data[['Finishing', 'Volleys', 'FKAccuracy',
                                 'ShotPower', 'LongShots', 'Penalties']].mean()).mean()))
```

```python
# Adding these categories to the data

fifa_data['Defend']= fifa_data.apply(defend, axis = 1)
fifa_data['General']= fifa_data.apply(general, axis= 1)
fifa_data['Mental'] = fifa_data.apply(mental, axis = 1)
fifa_data['Passing'] = fifa_data.apply(passing, axis = 1)
fifa_data['Mobility'] = fifa_data.apply(mobility, axis = 1)
fifa_data['Power'] = fifa_data.apply(power, axis = 1)
fifa_data['Rating']= fifa_data.apply(rating, axis = 1)
fifa_data['Shooting']= fifa_data.apply(shooting ,axis = 1)
```

```python
# lets check the Distribution of Scores of Different Skills

plt.rcParams['figure.figsize'] =(18,8)
plt.subplot(2,4,1)
sns.distplot(fifa_data['Defend'], color= 'red')
plt.grid()

plt.subplot(2,4,2)
sns.distplot(fifa_data['General'], color = 'green')
plt.grid()

plt.subplot(2,4,3)
sns.distplot(fifa_data['Mental'], color= 'black')
plt.grid()

plt.subplot(2,4,4)
sns.distplot(fifa_data['Passing'],color = 'yellow')
plt.grid()

plt.subplot(2,4,5)
sns.distplot(fifa_data['Mobility'], color = 'green')
plt.grid()

plt.subplot(2,4,6)
sns.distplot(fifa_data['Power'], color = 'black')
plt.grid()

plt.subplot(2,4,7)
sns.distplot(fifa_data['Shooting'], color = 'red')
plt.grid()

plt.subplot(2,4,8)
sns.distplot(fifa_data['Rating'], color = 'yellow')
plt.grid()

plt.title('Score Distributions for Different Abilities')
plt.show()
```
