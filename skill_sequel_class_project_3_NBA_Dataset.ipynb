```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
nba['Conference'] = encoder.fit_transform(nba['Conference'])
```

```python
nba['Role'] = encoder.fit_transform(nba['Role'])
```

```python
nba['Play'] = encoder.fit_transform(nba['Play'])
```

```python
nba
```

```python
# now i building a model
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = nba.drop(['Play','Season'], axis = 1)
```

```python
X
```

```python
y = nba['Play']
```

```python
X
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20,random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model1 = RandomForestClassifier()
```

```python
model1.fit(X_train,y_train)
```

```python
y_pred = model1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
# 98% accuracy without applying PCA
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 20)
pca_train = pca.fit_transform(X_train)
pca_test = pca.transform(X_test)
```

```python
model2 = RandomForestClassifier()
```

```python
model2.fit(pca_train,y_train)
```

```python
y_pred1 = model2.predict(pca_test)
```

```python
print(classification_report(y_test,y_pred1))
```

```python
# 98% accuracy after applying PCA
```
